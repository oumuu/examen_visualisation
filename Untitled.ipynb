```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score, accuracy_score, confusion_matrix
from imblearn.over_sampling import SMOTE
import warnings
warnings.filterwarnings('ignore')
datasets = {

    'abalone': {

        'original': 'abalone9-18.csv',

        'equilibre': 'equi-abalone9-18.csv'

    },

    'ecoli_147_vs_56': {

        'original': 'ecoli-0-1-4-7_vs_5-6.csv',

        'equilibre': 'equi-ecoli-0-1-4-7_vs_5-6.csv'

    },

    'ecoli4': {

        'original': 'ecoli4.csv',

        'equilibre': 'equi-ecoli4.csv'

    },

    'glass_016_vs_2': {

        'original': 'glass-0-1-6_vs_2.csv',

        'equilibre': 'equi-glass-0-1-6_vs_2.csv'

    },

    'glass_016_vs_5': {

        'original': 'glass-0-1-6_vs_5.csv',

        'equilibre': 'equi-glass-0-1-6_vs_5.csv'

    },

    'glass2': {

        'original': 'glass2.csv',

        'equilibre': 'equi-glass2.csv'

    },

    'glass4': {

        'original': 'glass4.csv',

        'equilibre': 'equi-glass4.csv'

    },

    'page_blocks': {

        'original': 'page-blocks-1-3_vs_4.csv',

        'equilibre': 'equi-page-blocks-1-3_vs_4.csv'

    },

    'shuttle_c0_vs_c4': {
    'original': 'shuttle-c0-vs-c4.csv',
    'equilibre': 'equi-shuttle-c0-vs-c4.csv'
    },
    'yeast_1_vs_7': {
        'original': 'yeast-1_vs_7.csv',
        'equilibre': 'equi-yeast-1_vs_7.csv'
    }

}

def load_dataset(name, version='original'):

    """Charge un dataset original ou équilibré"""

    filepath = datasets[name][version]

    df = pd.read_csv(filepath)

    # Recherche automatique de la colonne cible

    possible_targets = ['class', 'Class', 'target', 'Target', 'label', 'Label']

    target_col = None

    for col in possible_targets:

        if col in df.columns:

            target_col = col

            break

    # Sinon, dernière colonne

    if target_col is None:

        target_col = df.columns[-1]

        print(f"⚠️ {name} : colonne cible = '{target_col}'")

    return df, target_col
```

```python
print("\n=== QUESTION 1 : Visualisation du déséquilibre initial ===\n")

fig, axes = plt.subplots(2, 5, figsize=(20, 8))
axes = axes.flatten()

for idx, (name, filename) in enumerate(dataset_files.items()):
    df, target = load_dataset(name)
    counts = df[target].value_counts()
    
    axes[idx].bar(counts.index.astype(str), counts.values, color=['steelblue', 'coral'])
    axes[idx].set_title(f"{name}\n(minoritaire: {counts.min()} / majoritaire: {counts.max()})", fontsize=10)
    axes[idx].set_xlabel("Classe")
    axes[idx].set_ylabel("Fréquence")
    
    for i, v in enumerate(counts.values):
        axes[idx].text(i, v + max(counts)*0.02, str(v), ha='center', fontweight='bold')
    
    axes[idx].grid(axis='y', alpha=0.3)

plt.suptitle("Déséquilibre des classes - 10 datasets", fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------------------
# QUESTION 2 & 3 : Équilibrage (SMOTE) & re-visualisation
# -------------------------------------------------
print("\n=== QUESTION 3 : Visualisation après équilibrage (SMOTE) ===\n")

balanced_datasets = {}

fig2, axes2 = plt.subplots(2, 5, figsize=(20, 8))
axes2 = axes2.flatten()

for idx, (name, filename) in enumerate(dataset_files.items()):
    df, target = load_dataset(name)
    X = df.drop(columns=[target])
    y = df[target]
    
    # Application de SMOTE
    smote = SMOTE(random_state=42)
    X_res, y_res = smote.fit_resample(X, y)
    df_bal = pd.DataFrame(X_res, columns=X.columns)
    df_bal[target] = y_res
    balanced_datasets[name] = df_bal
    
    counts = df_bal[target].value_counts()
    
    axes2[idx].bar(counts.index.astype(str), counts.values, color=['lightgreen', 'lightblue'])
    axes2[idx].set_title(f"{name} (équilibré)", fontsize=10)
    axes2[idx].set_xlabel("Classe")
    axes2[idx].set_ylabel("Fréquence")
    
    for i, v in enumerate(counts.values):
        axes2[idx].text(i, v + 2, str(v), ha='center')

plt.suptitle("Datasets équilibrés par SMOTE", fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```

```python
print("\n=== QUESTIONS 6,7,8 : Comparaison F1, Accuracy, Matrice de confusion ===\n")
print("Dataset choisi : yeast-1_vs_7")

# Charger original et équilibré
df_orig, target = load_dataset('yeast_1_vs_7', 'original')
df_bal, _ = load_dataset('yeast_1_vs_7', 'equilibre')

X_orig = df_orig.drop(columns=[target])
y_orig = df_orig[target]
X_bal = df_bal.drop(columns=[target])
y_bal = df_bal[target]

# Split train/test
X_train_o, X_test_o, y_train_o, y_test_o = train_test_split(X_orig, y_orig, test_size=0.3, random_state=42)
X_train_b, X_test_b, y_train_b, y_test_b = train_test_split(X_bal, y_bal, test_size=0.3, random_state=42)

# Algorithmes
models = {
    'Random Forest': RandomForestClassifier(random_state=42),
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'SVM': SVC(random_state=42),
    'KNN': KNeighborsClassifier(n_neighbors=5),
    'Decision Tree': DecisionTreeClassifier(random_state=42)
}

results_f1 = {'Original': {}, 'Équilibré (WEKA)': {}}
results_acc = {'Original': {}, 'Équilibré (WEKA)': {}}

for name, model in models.items():
    # Original
    model.fit(X_train_o, y_train_o)
    y_pred_o = model.predict(X_test_o)
    results_f1['Original'][name] = f1_score(y_test_o, y_pred_o, average='weighted')
    results_acc['Original'][name] = accuracy_score(y_test_o, y_pred_o)
    
    # Équilibré
    model.fit(X_train_b, y_train_b)
    y_pred_b = model.predict(X_test_b)
    results_f1['Équilibré (WEKA)'][name] = f1_score(y_test_b, y_pred_b, average='weighted')
    results_acc['Équilibré (WEKA)'][name] = accuracy_score(y_test_b, y_pred_b)

# QUESTION 6 : Histogramme du F1-Measure
fig6, ax6 = plt.subplots(figsize=(12, 6))
x = np.arange(len(models))
width = 0.35

orig_f1 = list(results_f1['Original'].values())
bal_f1 = list(results_f1['Équilibré (WEKA)'].values())

ax6.bar(x - width/2, orig_f1, width, label='Original', color='coral', alpha=0.8)
ax6.bar(x + width/2, bal_f1, width, label='Équilibré (WEKA)', color='seagreen', alpha=0.8)

ax6.set_ylabel('F1-Score (weighted)', fontsize=12)
ax6.set_title('Comparaison du F1-Measure avant/après équilibrage (WEKA)', fontsize=14)
ax6.set_xticks(x)
ax6.set_xticklabels(results_f1['Original'].keys(), rotation=45, ha='right')
ax6.legend()
ax6.set_ylim(0, 1.05)
ax6.grid(axis='y', alpha=0.3)

for i, (o, b) in enumerate(zip(orig_f1, bal_f1)):
    ax6.text(i - width/2, o + 0.02, f'{o:.3f}', ha='center', fontsize=9)
    ax6.text(i + width/2, b + 0.02, f'{b:.3f}', ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
# QUESTION 7 : Diagramme en boîte (Boxplot) de l'Accuracy
print("\n--- Question 7 : Diagramme en boîte de l'Accuracy ---\n")
data_box = []
for model_name in results_acc['Original']:
    data_box.append({'Dataset': 'Original', 'Algorithme': model_name, 
                     'Accuracy': results_acc['Original'][model_name]})
    data_box.append({'Dataset': 'Équilibré (WEKA)', 'Algorithme': model_name, 
                     'Accuracy': results_acc['Équilibré (WEKA)'][model_name]})

df_box = pd.DataFrame(data_box)

#boxplot
data_to_plot = []
labels = []

for model in df_box['Algorithme'].unique():
    data_to_plot.append(df_box[df_box['Algorithme'] == model]['Accuracy'].values)
    labels.append(model)

plt.figure(figsize=(14,6))

box = plt.boxplot(data_to_plot, labels=labels, patch_artist=True)

# Couleurs alternées
colors = ['salmon', 'seagreen'] * len(labels)
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

plt.xticks(rotation=45)
plt.ylabel("Accuracy")
plt.title("Question 7 : Accuracy (Original vs Équilibré WEKA)")
plt.tight_layout()
plt.show()
```

```python
# QUESTION 8 : Matrice de confusion (Random Forest sur dataset équilibré)
print("\n--- Question 8 : Matrice de confusion (Random Forest - Dataset équilibré) ---\n")

best_model = RandomForestClassifier(random_state=42)
best_model.fit(X_train_b, y_train_b)
y_pred_best = best_model.predict(X_test_b)

cm = confusion_matrix(y_test_b, y_pred_best)

plt.figure(figsize=(7, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=np.unique(y_test_b), 
            yticklabels=np.unique(y_test_b))
plt.title(f"Matrice de confusion - Random Forest\n(Dataset yeast équilibré par WEKA)", fontsize=12)
plt.xlabel("Prédiction", fontsize=11)
plt.ylabel("Vérité terrain", fontsize=11)
plt.tight_layout()
plt.show()
```
